## Exploratory Data Analysis

### Research Question 1 
##### Over the years, who are the popular directors?

```python
# Importing data from newly cleaned dataset
df_dir = pd.read_csv('data_imdb_dir.csv')
df_dir.info()
```

```python
#Create basic histograms
#Initial exploration to see possible trends visually
df_dir.hist();
```

```python
# we take counts of releases for every director
pop_dir = df_dir['director'].value_counts()
pop_dir[:5]
```

This shows us the directors with maximum releases over the years but is not indicative of popularity. Hence, we try grouping by director names to gauge the popularity better.

```python
# save dataframe grouped by directors separately
# so that we can take consider every director's releases over the years together
# this is to see each directors efforts holistically
director_df = df_dir.groupby('director') 
dir_mean_df = director_df.mean()
dir_mean_df
```

```python
director_df.median()
```

```python
# we plot a histogram of each popularity garnered by directors
dir_mean_df['popularity'].hist();
plt.xlabel('Popularity')
plt.ylabel('Counts')
plt.title('Popularity Over the Years');
```

```python
# inspecting the statistics for the popularity of directors as a whole
#for thsi we take statistics for the dataset grouped by 'director' variable which was created in preceding steps
dir_mean_df['popularity'].describe()
```

```python
# directors with maximum popularity ratings
dir_mean_df['popularity'].nlargest(10) 
```

```python
dir_mean_df[880:900]#double check values
```

### Observations on Popularity:
- We have here the list of ten most popular directors.
- We also see that most directors score between 0 and 1 as per popularity. 
- The distribution is very skewed to the right.

## Exploratory Data Analysis

### Research Question 4
#### Is there a relation between popularity and revenue for directors?
Steps:
- To answer this question, we visualize the revenue and popularity variable from our dataset grouped over the director factor.

```python
# plotting popularity and revenue of directors
plt.scatter(x=dir_mean_df['popularity'], y=dir_mean_df['revenue']);
plt.xlabel('Popularity')
plt.ylabel('Revenue')
plt.title('Popularity vs Revenue');
```

```python
# computing correlation, if any, among variables in the dataset grouped by 'directors'
dir_mean_df.corr(method='pearson')
```

Observations:
- We see fromm our scatterplot and correlation that popularity and revenue for directors is positively correlated, but not  very significantly.
- But we do notice that for at lower values, the relation is significant.
- However, this is purely correlation and not to be inferred as causal.
- It does point us to a line of investigation that we can take up for detailed assessment.

## Limitations
- In the above analysis, we have worked with runtimes, popularity and revenues.
- Dataset is not exhaustive and hence, results or inferences are not absolute. i.e The dataset does not confirm that every release of every director is listed. Findings are limited to datset only.
- We restrict ourselves to only numerical values of revenues. Any variances in currencies, like dollars, rupees, Euros, etc. are not considered.
- In case of varying currencies [which is likely since the movies range from different regions], exchange rates or conversions or some sort of normalization would need to be applied.
- We also do not consider different revenue sources like movies ticket sales, audio rights sales, digital sales, braodcast sales, etc. and disregard the period and markets over which the revenues are collected. Older movies that are cast over television or re-released and might garner higher revenues are examples of limitations of the dataset and hence, this EDA.
- Vote counts are taken at face value. We are not inspecting the nature of votes, good or bad or any other range of values. Also, the period or channels in which these vote counts were gathered is not considered and will be significant if varying for different movies.
- The medium through which popularity was determined is unknown. This can impact the analysis as the limitations and bias, inherent while gauging audience response, will be present in the end values too.
- This analysis assumes that the same index and methods were employed for collecting popularity factors and counting votes for all movies. In the event that it is not so, the results might not hold true. [a possibility since movies are from different countries and languages]
- We are also considering body of work for every directors individual work. Collaborations are considered separate and only for unique set of directors. As stated at the beginnig, this analysis assumes collective preferences when directors work together.
- New and veteran directors are treated alike. Skew is possible due to directors having multiple releases in a single year or releases over many years over new directors or directors with fewer releases every year.
- While we did not have missing values for any of the factors under consideration, we acknowledge the presence of these limitations and assumptions in our analysis.
